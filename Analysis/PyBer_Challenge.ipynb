### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "../Resources/city_data.csv"
ride_data_to_load = "../Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
totalrides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
totalrides_by_type
```

```python
Type = pyber_data_df['type'].unique()
Type
```

```python
city_data_df
```

```python
# 2. Get the total drivers for each city type
totaldrivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
totaldrivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
totalfares_by_type_count = pyber_data_df.groupby(["type"]).count()["fare"]
totalfares_by_type_count
```

```python
totalfares_by_type_sum = pyber_data_df.groupby(["type"]).sum()["fare"]
totalfares_by_type_sum
```

```python
fare_per_ride = totalfares_by_type_sum / totalrides_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
fare_per_ride = totalfares_by_type_sum / totalrides_by_type
fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
fare_per_driver = totalfares_by_type_sum / totaldrivers_by_type
fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
summary_df = pd.DataFrame({
    'Total Rides': totalrides_by_type,
    'Total Drivers': totaldrivers_by_type,
    'Total Fares': totalfares_by_type_sum,
    'Average Fare Per Rider': fare_per_ride,
    'Average Fare Per Driver': fare_per_driver})
summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
summary_df.index.name = None
summary_df
```

```python
#  8. Format the columns.
summary_df["Total Fares"] = summary_df["Total Fares"].map("${:,.2f}".format)
summary_df["Average Fare Per Rider"] = summary_df["Average Fare Per Rider"].map("${:,.2f}".format)
summary_df["Average Fare Per Driver"] = summary_df["Average Fare Per Driver"].map("${:,.2f}".format)
summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
#pyber_data_df['date']=pd.to_datetime(pyber_data_df['date']).dt.date 
sum_fares_bydate = pyber_data_df.groupby(['type','date']).sum()["fare"]
#sum_fares_bydate_df.head()
sum_fares_bydate_df = pd.DataFrame(sum_fares_bydate)
sum_fares_bydate_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_fares_bydate_df = sum_fares_bydate_df.reset_index()
sum_fares_bydate_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_df = sum_fares_bydate_df.pivot(index='date', columns ='type', values='fare')
pivot_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
filterbydate_pivot = pivot_df.loc['2019-01-01':'2019-04-29']
#filterbydate_pivot_df = pivot_df.loc[:,'Day Of Year']='2019-01-01':'2019-04-29'
sum_fares_bydate_df = pd.DataFrame(filterbydate_pivot)
sum_fares_bydate_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
sum_fares_bydate_df.index = pd.to_datetime(sum_fares_bydate_df.index)
#filterbydate_pivot_df['date']=pd.to_datetime(filterbydate_pivot_df['date']).dt.date
sum_fares_bydate_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
sum_fares_bydate_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fare_df = sum_fares_bydate_df.resample('w').sum()
weekly_fare_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
weekly_fare_df.plot()
plt.ylabel("Fare($)")
plt.xlabel("Date")
plt.title("Fare By City Type")
plt.show()
```
